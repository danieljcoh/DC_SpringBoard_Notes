# Using `iloc()`

- `iloc()` allows us to filter through our data using integers instead of labels or text

- This means: In our df, DataFrame, go to the "4th" row (because it's 0 indexed) and print out that entire row.
- This will include all the labels and data pieces for a singular row of data

```python
df.iloc[3]
```

- We can specify further even just singular piece of data from that row.
- This could mean, everything from the "4th" row but only the 1st column, which depending on our data labels in the columns, would only print out one piece of data corresponding to a singular data row

```python
print("latitude: " + str(df.iloc[3, 1]))
print("housing_median_age: " + str(df.iloc[3, 2]))
```

### Slicing our DataFrame

- We can slice data from rows to consolidate specific data we are looking to retrieve
- Here we sliced just rows 10 and 11
- As you can see the slicing command is **exclusive**

```python
df.iloc[10:12]
```

## Using `apply()`

- Here we are going to make a new column to the DataFrame but we are doing so for every row of data by using the `apply()` method and taking the data from "total_bedrooms" and see if it meets a certain criteria
- The lambda function here lets us apply a certain criteria to each piece of data in the new cell based on the old data from "total_bedrooms"

```python
df["Happiness"] = df["total_bedrooms"].apply(lambda x: x > 500)
```
